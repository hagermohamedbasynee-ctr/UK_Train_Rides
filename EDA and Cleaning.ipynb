* Cleaning the Null values in each column

```python
# Railcard column -> Whether the passenger is a National Railcard holder (Adult, Senior, or Disabled) or not (None).
df["Railcard"] = df["Railcard"].fillna("No Railcard")
```

```python
df.isnull().sum()
```

```python
df['Reason for Delay'].unique()
```

```python
# Unify inconsistent Reason for Delay values
delay_map = {
    "Weather": "Weather Conditions",
    "Signal failure": "Signal Failure",
    "Staffing": "Staff Shortage",
}
df["Reason for Delay"] = df["Reason for Delay"].replace(delay_map)
print("\nReason for Delay after unifying:")
print(df["Reason for Delay"].value_counts())
```

```python
# Cleaning the null values from Reason for Delay
df["Reason for Delay"] = df.apply(
    lambda row: "No Delay" if row["Journey Status"] == "On Time" and pd.isnull(row["Reason for Delay"])
    else "Unknown" if pd.isnull(row["Reason for Delay"])
    else row["Reason for Delay"],
    axis=1
)
```

```python
print(df["Reason for Delay"].value_counts())
```

```python
df.isnull().sum()
```

```python
# Actual Arrival Time
df["Actual Arrival Time"] = df.apply(
    lambda row: "Cancelled" if row["Journey Status"] == "Cancelled" and pd.isnull(row["Actual Arrival Time"])
    else "Unknown" if pd.isnull(row["Actual Arrival Time"])
    else row["Actual Arrival Time"],
    axis=1
)
```
